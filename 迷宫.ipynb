```python
from random import randint

def mark(maze,pos):
    '''
    标记函数，用来标记历史走过的位置
    maze：一个m*n的二维矩阵迷宫
    pos: 当前需要标记的位置坐标 pos=(x,y),x=pos[0],y=pos[1]
    '''
    maze[pos[0]][pos[1]]=2 #将走过的位置标记为2

def move(maze,pos):
    '''
    移动函数：用来测试当前位置是否可继续移动，移动条件为当前位置为0
     maze：一个m*n的二维矩阵迷宫
    pos: 当前需要标记的位置坐标 pos=(x,y),x=pos[0],y=pos[1]
    return：bool类型
    '''
    return maze[pos[0]][pos[1]]==0

move_path=[] #记录能成功到达出口的移动路径坐标
path_direction=[] #记录能成功到达出口的移动路径的方向

def find_path(maze,start,end):
    '''
    路径查找函数
    maze：一个m*n的二维矩阵迷宫
    start: 起始点位置坐标，start=(1,1)
    end: 终点坐标，end=(m,n)
    reutrn bool类型
    '''
    mark(maze,start)  #将起始位置标记
    if start==end:  #路径查找终止条件是到达终点
        move_path.append(start)
        return True
    
    #未达终点，存在4种可能的移动方向，即向上(-1,0),下(1,0),左(0,-1),右(0,1)
    move_direction=[(-1,0),(1,0),(0,-1),(0,1)]
    direction=["↑","↓","←","→"]
    for i in range(4): #遍历4种可能的方向
        next_start=(start[0]+move_direction[i][0],start[1]+move_direction[i][1]) #下一个坐标起点
        if move(maze,next_start):
            if find_path(maze,next_start,end):
                #这里之所以仍然添加起始坐标是因为当查询下一个位置就是终点或者可到达终点时记录此时位置
                move_path.append(start)
                path_direction.append(direction[i]) #记录路径方向
                return True
    return False

def gen_maze(m,n):
    '''
    生成随机迷宫阵列
    '''
    m+=2
    n+=2  #m和n均+2是为了构造外层的1
    maze=[[1 for i in range(n)] for j in range(m)] #初始化大小为m*n,值全为1的二维矩阵
    for x in range(1,m-1):
        for y in range(1,n-1):
            '''
            这里x,y取值范围为x属于1，m-1;y为1，n-1 是因为我们让该迷宫的最外层全为1
            '''
            if (x==1 and y==1) or (x==m-2 and y==n-2):
                maze[x][y]=0 #起点必须为0
            else:
                maze[x][y]=randint(0,1)  #在最外层全为1的情况下，随机取值0,1
    return maze

def print_maze(maze,text="原始迷宫为：",end1=" ",end2="/n/n",xs=0,xe=0,ys=0,ye=0):
    '''
    输出迷宫矩阵
    '''
    print(text)
    n,m=len(maze[0]),len(maze)
    for x in range(xs,m-xe):
        for y in range(ys,n-ye):
            print(maze[x][y],end=end1)
        print(end=end2)
    
    
def path_maze(maze,directions_map):
    '''
    生成带有移动路径的迷宫矩阵
    maze：一个m*n的二维矩阵迷宫
    directions_map:一个记录移动方向坐标的字典，有↑，↓，←，→四个元素
    return：path_maze
    '''
    n,m=len(maze[0]),len(maze)
    for x in range(1,m-1):
        for y in range(1,n-1):
            maze[x][y]=maze[x][y] if maze[x][y] !=2 else 0 #将标记为2还原为0
            
    for x in range(m):
        for i in range(1,2*n-1,2):
            maze[x].insert(i,"    ") #初始化maze,在每两个元素间插入占位符"   "
            
    for x in range(1,2*m-1,2):
        maze.insert(x,[" ","    "]*(n-1)+[" "])
        
    for direction in directions_map:
        for directions_position in directions_map[direction]:
            i,j=directions_position
            i=2*i
            j=2*j
            if direction=="↑":
                maze[i-1][j]="↑"
            if direction=="↓":
                maze[i+1][j]="↓"
            if direction=="←":
                maze[i][j]="←"
            if direction=="→":
                maze[i][j+1]=" → "
    return maze
```
